```python
'''
#!/bin/sh
# =================================== #
# 功能 ：根据目标的指定条件进行删除，再调用 进行加载
#
# =================================== #

usage(){
    
    echo "   ===================="
    echo "   Usage:"
    echo "   sh $0  -q 日期   -f filename "
    echo "   参数:"
    echo "       -q 日期   "
 
    echo "       -f datefilename  H010_Load所用到的数据文件"
 
    exit -1
}

VERSION="V1.0"
CMD_NAME=$0
dt=`date "+%Y%m%d"`
LogFile=${HOME}/han/log/fhtsDup${dt}.log
LogDir=`dirname ${LogFile}`:x
[ ! -d ${LogDir} ] && mkdir -p ${LogDir}

#记录日志
shlog()
{
    echo "【"`date "+%Y%m%d %H:%M:%S"`"】$* ">> ${LogFile}
}
shlogdsp()
{
    echo "【"`date "+%Y%m%d %H:%M:%S"`"】$CMD_NAME ":" $* ">> ${LogFile}
    echo "【"`date "+%Y%m%d %H:%M:%S"`"】$* "
}
shlogdsp "VERSION:${VERSION}"


if [ $# -lt 4 ]
then
    shlogdsp "参数有误,请检查......"
    usage
fi
__params=" "
while [ 1 ]
do
     
    if [ "$1" = "-f" ]
    then
        shift 1
        _filename=$1
    fi
    if [ "$1" = "-q" ]
    then
        shift 1
        _date=$1
    fi
     shift 1
    if [ $# -eq 0 ]
    then
        break
    fi
done
shlogdsp "${filename} ${_date}"
 
#判断参数是否正确
 
if [ "${_filename}" = "" ]
then 
     shlogdsp "ERROR:参数 _filename 数据文件 为空"
     usage
fi
 
if [ "${_date}" = "" ]
then
     shlogdsp "ERROR:参数 _date 日期数据格式匹配 ?为空"
     usage
fi

 
shlogdsp "开始处理数据..."
if [ ! -f ${_filename} ]
then 
exit -4
fi
while [ 1 ] 
do 
_systime=`date +%s`
_filetime=`stat -c %Y ${_filename}`
_diff=`expr ${_systime} - ${_filetime} `
echo  ${_diff}
if [ ${_diff} -gt 100 ] 
then
 break
fi
sleep 10
done 

cp ${_filename} ${_filename}.bak
sort -u -t\| -k4,4 -k6,6 -k8,8 -k9,9 -k10,10 -k11,11 ${_filename}.bak > tmp${_date}
mv tmp${_date} ${_filename}
ls -l ${_filename}
sleep 10


ret=$?
if [ $ret -ne 0 ]
then
    shlogdsp "数据加载失败,程序退出."
        exit -4
else
        shlogdsp "数据加载完成."
        
fi

'''
```
